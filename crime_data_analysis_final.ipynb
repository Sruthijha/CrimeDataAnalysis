# Dataset Description

The dataset estimated_crimes_1979_2023.csv was sourced from the FBI Crime Data Explorer and forms the foundation of our analysis on crime trends in the United States. Spanning the years 1979 through 2023, it contains detailed records of both violent and property crimes reported annually at the national and state levels. Each record includes population estimates and crime counts for categories such as homicide, aggravated assault, robbery, burglary, larceny, and motor vehicle theft. To facilitate accurate temporal and geographic comparisons, crime counts were normalized to rates per 100,000 population. This observational dataset provides an essential historical perspective on public safety, enabling our study to identify long-term patterns, regional disparities, and key anomalies such as the crime spikes in the early 1990s and disruptions during the COVID-19 pandemic. Through data cleaning, transformation, and rigorous statistical analysis, this dataset supports evidence-based policy insights aimed at improving crime prevention and resource allocation across the United States.

# 3. Loading the Dataset

```r
crime_data = read.csv("estimated_crimes_1979_2023.csv")

head(crime_data)
```

# 4. Overview of Dataset Attributes

```r
colnames(crime_data)
```

Understanding each attribute in the dataset

1. year: The calender year the date corresponds to (1973, 2023)
2. state_abbr: Abbrevation to U.S. state
3. state_name: Full name of U.S. state
4. population: Estimated population of the state or country for the respective year
5. violent_crime: Total no.of reported violent crimes (sum of homicide, rape, robbery, and aggravated assault)
6. Homicide: Number of reported homicides (intentional killing of another person)
7. rape_legacy: Number of reported rape cases using the legacy definition (prior to the FBI’s updated definition)
8. rape_revised: Number of reported rape cases using the revised definition (introduced in 2013).
9. robbery: Number of reported robberies (theft involving force or threat of force).
10. aggravated_assault: Number of reported aggravated assault cases (attacks intended to cause severe bodily harm)
11. property_crime: Total number of reported property crimes (sum of burglary, larceny, and motor vehicle theft)
12. burglary: Number of reported burglaries (unlawful entry into a structure to commit theft or felony)
13. larceny: Number of reported larcenies (theft of personal property without force).
14. motor_vehicle_theft: Number of reported motor vehicle thefts (stealing or attempting to steal vehicles)
15. caveats: Additional notes or disclaimers about the data. For instance, changes in definitions, underreporting, or data anomalies

# 5. Handling Data Type Discrepancies

- Converts year to integer: Ensures proper formatting for chronological analysis.
- Handles numeric columns: Removes commas (if any) and converts population and crime-related metrics to numeric.
- Converts state-related columns to character: Ensures these columns are treated as strings.

```r
crime_data = crime_data %>%
  mutate(year = as.integer(year))
```

```r
# Converting population and crime-related columns to numeric

numeric_columns = c(
  "population", "violent_crime", "homicide", "rape_legacy",
  "rape_revised", "robbery", "aggravated_assault",
  "property_crime", "burglary", "larceny", "motor_vehicle_theft"
)

crime_data = crime_data %>%
  mutate(across(all_of(numeric_columns), ~ as.numeric(gsub(",", "", .))))
```

```r
# Converting state abbreviation and state name to character

crime_data = crime_data %>%
  mutate(
    state_abbr = as.character(state_abbr),
    state_name = as.character(state_name)
  )
```

We inspect the dataset for missing values and take steps to replace empty strings with NA, drop irrelevant columns, and impute where appropriate.

```r
# Dropping the column if it has minimal or irrelevant data

if (empty_caveats_percentage > 90) {
  crime_data = crime_data %>% select(-caveats)
  print("'caveats' column dropped due to irrelevance.")
} else {
  print("'caveats' column retained for reference.")
}
```

### Handling all empty strings in the Dataset

```r
# Replacing all empty strings in the dataset with NA

crime_data[crime_data == ""] = NA

missing_summary = colSums(is.na(crime_data))
print("Missing values after replacing empty strings:")
print(missing_summary)
```

# 7. Seperating the National and State related Crime Data

$\text{If the state name is empty then that column represents the national level data. So its good to seperate this data to perform timely analysis on national-level trends and state-wise}$
$\text{analysis on state-specific data.}$

```r
# Identifying nation-level records

national_data = crime_data[(
  is.na(crime_data$state_name) |
  crime_data$state_name == "" |
  crime_data$state_name == "United States Total"
) & is.na(crime_data$state_abbr), ]
```

```r
# Removing these records from the original dataset

state_data = crime_data[!(
  (is.na(crime_data$state_name) |
   crime_data$state_name == "" |
   crime_data$state_name == "United States Total") &
  is.na(crime_data$state_abbr)
), ]
```

```r
# Replacing state_abbr and state_name in national_data with 'USA'

national_data$state_abbr = "USA"
national_data$state_name = "USA"

head(national_data[c("state_abbr", "state_name")])
```

Understanding more on rape legacy and rape revised attributes to handle missing values

1. Rape Legacy:
This column contains data collected using the legacy definition of rape.
It was used prior to 2013 and includes only forcible rape of a female.

2. Rape Revised:
This column contains data collected using the revised definition introduced in 2013 by the FBI.
The revised definition is broader and includes:
Rape of both males and females.
Penetration without consent, regardless of gender or force.

Years Pre-2013:
Only use rape_legacy since rape_revised will likely be missing.
Years Post-2013:
Only use rape_revised since it provides more comprehensive data.
Transition Period (Around 2013):
Use rape_revised where available, as it’s the more accurate measure.

If both columns are populated for some records:

Prefer rape_revised for analysis, as it is more inclusive and reflects modern definitions.
Optionally, create a new column that prioritizes rape_revised and uses rape_legacy as a fallback.

```r
# Creating a unified column for rape counts

national_data$rape = ifelse(!is.na(national_data$rape_revised),
                             national_data$rape_revised,
                             national_data$rape_legacy)
```

# 9. Data Normalization

Normalization is a crucial step to ensure that crime data is comparable across states and years. By calculating crime rates per 100,000 population, we standardize the data, allowing for fair comparisons regardless of population differences.

```r
crime_columns = c(
  'violent_crime', 'homicide', 'rape',
  'robbery', 'aggravated_assault', 'property_crime',
  'burglary', 'larceny', 'motor_vehicle_theft'
)

# Normalizing national data

for (col in crime_columns) {
  national_data[[paste0(col, "_rate")]] = (national_data[[col]] / national_data$population) * 100000
}

# Normalizing state data

for (col in crime_columns) {
  state_data[[paste0(col, "_rate")]] = (state_data[[col]] / state_data$population) * 100000
}

head(national_data)
head(state_data)
```

### highest and lowest crime rates for each crime type

```r
# Find the year with the highest and lowest crime rates for each crime type

key_metrics = lapply(crime_rate_columns, function(crime) {
  data.frame(
    Crime_Type = crime,
    Max_Year = national_data$year[which.max(national_data[[crime]])],
    Max_Value = max(national_data[[crime]], na.rm = TRUE),
    Min_Year = national_data$year[which.min(national_data[[crime]])],
    Min_Value = min(national_data[[crime]], na.rm = TRUE)
  )
})

# Combine results into a single data frame
key_metrics_df = do.call(rbind, key_metrics)
print(key_metrics_df)
```

### Year-over-Year (YoY) Change Analysis

```r
# List of crime rate columns
crime_rate_columns = grep("_rate$", names(national_data), value = TRUE)

# Calculating Year-over-Year (YoY) changes
national_data_yoy = national_data %>%
  mutate(across(all_of(crime_rate_columns), ~ . - lag(.), .names = "yoy_{.col}"))

# Finding the years with the largest increases and decreases for each crime type
yoy_summary = lapply(crime_rate_columns, function(crime) {
  yoy_values = national_data_yoy[[paste0("yoy_", crime)]]
  valid_years = national_data_yoy$year[!is.na(yoy_values)]
  valid_values = yoy_values[!is.na(yoy_values)]

  data.frame(
    Crime_Type = crime,
    Max_Year = valid_years[which.max(valid_values)],
    Max_Change = max(valid_values),
    Min_Year = valid_years[which.min(valid_values)],
    Min_Change = min(valid_values)
  )
})

yoy_summary_df = do.call(rbind, yoy_summary)

print(yoy_summary_df)
```


The analysis of year-over-year (YoY) changes in crime rates reveals distinct trends across decades and crime categories. The 1980s and early 1990s saw significant increases in crime rates, particularly in property crimes and violent crimes, with peaks in 1980 and 1990. However, the mid-to-late 1990s marked substantial declines, reflecting the impact of improved law enforcement practices and socioeconomic changes. In recent years, 2021 stands out with notable increases in violent crime rates and shifts in crime proportions, likely influenced by the societal disruptions of the COVID-19 pandemic. By 2022 and 2023, certain crime categories, including property crimes and larceny, continued to decline, signaling ongoing improvements in crime prevention measures. These findings highlight the evolving nature of crime trends and the influence of historical and contemporary factors on public safety.

# 13. Hypothesis Testing

### Shapiro-Wilk Test: Normality Check:

We first test whether the violent crime rates in the Northeast and West regions follow a normal distribution.

- **Null Hypothesis (H₀):** The data is normally distributed.
- **Alternative Hypothesis (H₁):** The data is not normally distributed.

```r
# Subset data for Northeast and West regions
northeast = state_data %>% filter(region == "Northeast") %>% pull(violent_crime_rate)
west = state_data %>% filter(region == "West") %>% pull(violent_crime_rate)

#Check Normality using Shapiro-Wilk Test
shapiro_northeast = shapiro.test(northeast)
shapiro_west = shapiro.test(west)

print("Shapiro-Wilk Test for Northeast:")
print(shapiro_northeast)

print("Shapiro-Wilk Test for West:")
print(shapiro_west)
```

The Shapiro-Wilk test results show that the violent crime rates for both the Northeast and West regions do not follow a normal distribution. In the Northeast, the test statistic is 0.80292 with a p-value of 3.46e-12, and in the West, the test statistic is 0.84768 with a p-value of 1.709e-10. Since both p-values are significantly lower than 0.05, the null hypothesis of normality is rejected for both regions. This indicates that alternative methods, such as non-parametric tests, may be more appropriate for analyzing violent crime rates in these regions.

### Levene’s Test: Homogeneity of Variance

Before running a t-test or ANOVA, we must check whether the variances of the two groups are equal.

- **Null Hypothesis (H₀):** The variances in violent crime rates are equal between the Northeast and West.
- **Alternative Hypothesis (H₁):** The variances are not equal between the two regions.

```r
levene_test = leveneTest(violent_crime_rate ~ region,
                          data = state_data %>% filter(region %in% c("Northeast", "West")))

levene_p_value = levene_test$`Pr(>F)`[1]
print("Levene's Test for Homogeneity of Variances:")
print(levene_test)
```

The Levene's Test for Homogeneity of Variances evaluates whether the variances of violent crime rates are equal between the Northeast and West regions. The test yielded an F-value of 0.1028 with a p-value of 0.7487, indicating that the variances are not significantly different. Since the p-value is much greater than the common threshold of 0.05, we fail to reject the null hypothesis of equal variances. This suggests that the assumption of homogeneity of variances holds for these two regions, making parametric methods, such as the t-test, appropriate for comparing their means.

### Z-test

Based on the Shapiro-Wilk test results, the data for some regions, such as Northeast and West, do not follow a normal distribution (p<0.05). The z-test assumes that the data is normally distributed, making it unsuitable for this dataset.

### T-test

### Independent Two-Sample t-Test

We compare the means of violent crime rates in the Northeast and West using a t-test.

- **Null Hypothesis (H₀):** There is no significant difference in mean violent crime rates between the Northeast and West.
- **Alternative Hypothesis (H₁):** There is a significant difference in mean violent crime rates between the Northeast and West.

```r
t_test_result = t.test(northeast, west,
                        alternative = "two.sided",
                        var.equal = levene_p_value > 0.05)

print("Two-Sample t-Test Result:")
print(t_test_result)

#Visualize the Data
ggplot(state_data %>% filter(region %in% c("Northeast", "West")),
       aes(x = region, y = violent_crime_rate, fill = region)) +
  geom_boxplot() +
  ggtitle("Comparison of Violent Crime Rates: Northeast vs West") +
  xlab("Region") +
  ylab("Violent Crime Rate (per 100,000)") +
  theme_minimal()
```

### Chi-Square Test of Independence

We examine whether the type of crime is independent of region.

- **Null Hypothesis (H₀):** Crime type distribution is independent of the region.
- **Alternative Hypothesis (H₁):** Crime type distribution depends on the region.


```r
# Testing the Association Between Crime Type and Region
# Create a contingency table: Count occurrences of crime types by region
crime_table = state_data %>%
  select(region, violent_crime, property_crime, burglary) %>%
  pivot_longer(cols = c(violent_crime, property_crime, burglary),
               names_to = "crime_type",
               values_to = "crime_count") %>%
  group_by(region, crime_type) %>%
  summarize(total_crimes = sum(crime_count, na.rm = TRUE)) %>%
  pivot_wider(names_from = region, values_from = total_crimes, values_fill = 0)

# Converting to a matrix for the chi-square test
crime_matrix = as.matrix(crime_table[-1])

# Performing the chi-square test
chi_sq_result = chisq.test(crime_matrix)

print("Chi-Square Test Results:")
print(chi_sq_result)
```

# 14. Boot strapping

It's especially useful when:

 - The sample size is small.
 - Normality assumptions are violated.
 - You need confidence intervals for robust estimations.

Bootstrapping is a non-parametric resampling technique that estimates the sampling distribution of a statistic (e.g., mean, median, variance) by repeatedly drawing samples with replacement from the original dataset.Bootstrapping is particularly useful for our analysis as it allows us to estimate confidence intervals (CIs) for key statistics, such as the mean or median of crime rates across regions or years, even when normality assumptions are not met. For example, we can calculate the 95% CI for the mean violent crime rate in the Northeast to better understand the true population mean. It also enables us to compare CIs between groups, like the Northeast and West, to identify significant differences without relying on parametric assumptions. This is especially helpful since the Shapiro-Wilk test indicated that our data is not normally distributed.

### Bootstrapping Confidence Intervals

We use bootstrapping to estimate confidence intervals for mean violent crime rates in different regions.

- This approach avoids normality assumptions and is useful when the data is skewed.

### Computing Bootstrap Confidence Interval for a Statistic (Mean)

```r
bootstrap_mean = function(data, indices) {
  mean(data[indices])
}

# Subset violent crime data for the Northeast region
northeast_data = state_data %>% filter(region == "Northeast") %>% pull(violent_crime_rate)

set.seed(123)
boot_result = boot(data = northeast_data, statistic = bootstrap_mean, R = 1000)

print(boot_result)

boot_ci = boot.ci(boot_result, type = "perc")
print(boot_ci)
```

### Comparing Confidence Intervals Across Regions

```r
bootstrap_by_region = function(region_name) {
  region_data = state_data %>% filter(region == region_name) %>% pull(violent_crime_rate)
  boot(data = region_data, statistic = bootstrap_mean, R = 1000)
}

# Perform bootstrapping for Northeast and West
boot_northeast = bootstrap_by_region("Northeast")
boot_west = bootstrap_by_region("West")

# Calculate confidence intervals
ci_northeast = boot.ci(boot_northeast, type = "perc")
ci_west = boot.ci(boot_west, type = "perc")

print("Northeast CI:")
print(ci_northeast)

print("West CI:")
print(ci_west)
```

### Visualizing Bootstrap Results

The histogram displayed represents the distribution of the dataset under analysis, showcasing a roughly bell-shaped curve indicative of a near-normal distribution. # 15. Outlier Detection and Handling

Outliers are extreme values that differ significantly from other observations in the dataset. They can distort statistical measures like the mean and inflate variability, potentially leading to misleading insights. Outliers in crime rates can significantly skew averages and impact statistical analyses such as hypothesis testing, bootstrapping, and regression results. In our dataset, the objective is to identify outliers in violent crime rates or other crime categories across states and regions. By analyzing and addressing these outliers, we can assess their influence on key statistical measures, such as the mean and median crime rates, and determine whether removing or adjusting them results in more robust and reliable insights.

Techniques to Identify Outliers:

- Z-Score Method
- Interquartile Range (IQR) Method
- Boxplot Visualization

### Z-Score Method

- Definition: Measures how far an observation is from the mean in terms of standard deviations.
- Rule of Thumb: Observations with Z-scores >3 or <−3 are considered outliers.

This method can't be used in our analysis

### Interquartile Range (IQR) Method

- Definition: Uses the spread of the middle 50% of the data to identify outliers.
- Rule of Thumb: Outliers are values outside [Q1−1.5×IQR, Q3+1.5×IQR].

### Outlier Identification

We use the IQR method to identify extreme values in the violent crime rate.

- Outliers can distort statistical tests.
- Since data is not normally distributed, we avoid Z-score methods.

```r
# Shapiro-Wilk Test
shapiro_test = shapiro.test(state_data$violent_crime_rate)
print(shapiro_test)

# Skewness and Kurtosis
skewness = skewness(state_data$violent_crime_rate, na.rm = TRUE)
kurtosis = kurtosis(state_data$violent_crime_rate, na.rm = TRUE)

print(paste("Skewness:", skewness))
print(paste("Kurtosis:", kurtosis))
```

The data for violent crime rates is not normally distributed, right-skewed, and has heavy tails.

These characteristics suggest that:
- The Z-score method may not be ideal for detecting outliers since it assumes normality.
- The IQR method is more appropriate for handling outliers in this case because it is robust to skewed data and heavy tails.

```r
# Calculating IQR for violent crime rates
Q1 = quantile(state_data$violent_crime_rate, 0.25, na.rm = TRUE)
Q3 = quantile(state_data$violent_crime_rate, 0.75, na.rm = TRUE)
IQR = Q3 - Q1

state_data = state_data %>%
  mutate(outlier_iqr = ifelse(violent_crime_rate < (Q1 - 1.5 * IQR) |
                              violent_crime_rate > (Q3 + 1.5 * IQR), TRUE, FALSE))

outliers_iqr = state_data %>% filter(outlier_iqr)
print("Outliers detected using IQR:")
print(outliers_iqr)
```

## 17. Hypothesis Testing with Randomization (Permutation Test)

We use a permutation test to evaluate whether the mean violent crime rate differs significantly between the Northeast and West.

- **Null Hypothesis (H₀):** There is no difference in the mean violent crime rates between the Northeast and West regions.
- **Alternative Hypothesis (H₁):** There is a significant difference in mean violent crime rates between the two regions.

This test does not rely on normality assumptions, making it a non-parametric alternative to the t-test.

```r
set.seed(123)
obs_diff = mean(northeast) - mean(west)
combined = c(northeast, west)
n = length(northeast)

perm_diffs = replicate(5000, {
  shuffled = sample(combined)
  mean(shuffled[1:n]) - mean(shuffled[(n + 1):length(combined)])
})

p_val_perm = mean(abs(perm_diffs) >= abs(obs_diff))

hist(perm_diffs, main = "Permutation Test: Northeast vs West",
     xlab = "Difference in Means", col = "lightblue", breaks = 30)
abline(v = obs_diff, col = "red", lwd = 2)
```

The histogram illustrates the results of a permutation test conducted to compare the mean violent crime rates between the Northeast and West regions of the United States. The distribution represents 5,000 simulated differences in means obtained by randomly reallocating regional labels (Northeast and West) across the pooled dataset. This process generates a null distribution assuming no true difference between the regions. The red vertical line marks the observed difference in means from the actual data. Since this observed value lies well within the center of the null distribution — and not in the extreme tails — it suggests that the difference in violent crime rates is not statistically significant. This visual reinforces the earlier t-test conclusion, supporting the hypothesis that any observed difference is likely due to chance rather than a meaningful regional disparity.

## 21. ANOVA (Comparing Means of Violent Crime Rates Across Regions)

We perform an Analysis of Variance (ANOVA) to test whether the mean violent crime rates differ across three regions: Northeast, West, and Other.

- **Null Hypothesis (H₀):** The mean violent crime rates are equal across all regions.
- **Alternative Hypothesis (H₁):** At least one region has a different mean violent crime rate.

### Assumptions:
- Observations are independent (each record corresponds to a unique state-year).
- Violent crime rates are normally distributed within each region (checked via Shapiro-Wilk Test).
- Homogeneity of variances across regions (verified using Levene’s Test).


```r
anova_result = aov(violent_crime_rate ~ region, data = state_data)
summary(anova_result)
```

This is the ANOVA (Analysis of Variance) output evaluating whether the violent crime rate differs significantly across regions.

- The independent variable is region, which has 3 groups (degrees of freedom = 2).

- The dependent variable is violent_crime_rate.

- The F-statistic is 4.774, with a p-value = 0.00853, which is statistically significant at the 1% level (denoted by **).

This means:

There is a statistically significant difference in the mean violent crime rates among at least two of the regions. In other words, region has a meaningful effect on violent crime rates.

### Tukey’s HSD Post-Hoc Test

Following a significant ANOVA result, we apply Tukey's Honest Significant Difference test to identify which specific regional pairs differ significantly in their mean violent crime rates.


```r
# Running ANOVA model again
anova_model = aov(violent_crime_rate ~ region, data = state_data)

# Applying Tukey's HSD test
tukey_result = TukeyHSD(anova_model)
print(tukey_result)
```

The Tukey HSD post-hoc test was conducted following a significant ANOVA result to identify which regional pairs differ significantly in their mean violent crime rates. The results reveal that the “Other” region has a significantly lower mean violent crime rate than the Northeast (mean difference = -70.51, p = 0.013), with a 95% confidence interval ranging from -129.10 to -11.92.

In contrast, the differences between the West and Northeast (p = 0.567) and between the West and Other (p = 0.325) are not statistically significant, as their confidence intervals include zero. These findings suggest that the primary driver of the overall regional difference in violent crime rates stems from the disparity between the Northeast and Other regions, while the West shares more similar rates with both.

This analysis supports a region-specific approach to policy and crime prevention, particularly targeting the elevated rates observed in the Northeast.

## 22. Generalized Linear Modeling (GLM - Poisson)

```r
# Fit GLM using Poisson distribution
glm_model = glm(violent_crime ~ year + population,
                 data = national_data,
                 family = poisson())

# Summary of GLM (optional but useful)
summary(glm_model)
```

**Poisson Generalized Linear Model (GLM) Interpretation**

A Poisson GLM was fitted to model violent crime counts using year and population as predictors. The results show that both predictors are highly statistically significant (p < 2e-16):

- The coefficient for year is positive (0.05022), indicating that violent crime counts tend to increase over time, after accounting for population changes.

- The coefficient for population is negative (-2.006e-08), suggesting that, controlling for time, states with larger populations are associated with lower overall violent crime counts. This may reflect the effect of improved infrastructure, policing efficiency, or per-capita dilution in more populated regions.

Despite the statistical significance of both predictors, the model fit is poor. The residual deviance (913,418) remains extremely high, and the AIC (914,080) is substantially worse than those from alternative models. This suggests that the Poisson distribution does not adequately capture the structure of the data, reinforcing that a more flexible model—like the Generalized Additive Model (GAM)—is more appropriate for analyzing violent crime trends over time.

## 23. Generalized Additive Model (GAM)

To model the long-term trend of violent crime in the United States, we use a GAM with a smooth spline on year. This model captures nonlinear temporal patterns better than linear models.

### Why GAM?
- It allows flexible modeling of year-over-year changes.
- It avoids rigid assumptions of linearity.
- It is evaluated using adjusted R², AIC, and deviance explained.


```r
if (!require("mgcv")) install.packages("mgcv")
library(mgcv)

# Fit GAM with smooth term for year
gam_model = gam(violent_crime_rate ~ s(year),
                 data = national_data)

# Summary of GAM
summary(gam_model)
```

**Generalized Additive Model (GAM) Interpretation**

A Generalized Additive Model (GAM) was used to model the violent crime rate as a smooth, nonlinear function of year, using a Gaussian distribution with an identity link function.

-The intercept estimate is 531.74 (p < 2e-16), representing the average baseline violent crime rate when the smooth term is centered. The smooth term for year is highly significant (p < 2e-16), with an effective degrees of freedom (edf) of 8.65, indicating a complex, nonlinear trend rather than a straight-line relationship.

The model demonstrates an excellent fit to the data:

- Adjusted R² = 0.982, meaning the model explains 98.2% of the variance in violent crime rates.

- Deviance explained = 98.6%, confirming a near-perfect match to the data’s structure.

- The Generalized Cross Validation (GCV) score of 333.05 and low scale estimate of 254.68 further support model reliability.

These results confirm that the GAM is the most appropriate model for capturing the historical trend of violent crime in the U.S. Its ability to flexibly model year-to-year changes significantly outperforms linear and Poisson-based models, making it ideal for both interpretation and forecasting.

### Plotting the GAM smooth term:

```r
plot(gam_model, se = TRUE, col = "blue",
     main = "GAM: Smoothed Trend of Violent Crime Rate Over Time",
     xlab = "Year", ylab = "Violent Crime Rate")
```

The plot titled "GAM: Smoothed Trend of Violent Crime Rate Over Time" illustrates the nonlinear relationship between year and violent crime rate, as modeled by the Generalized Additive Model (GAM). The solid blue line represents the estimated trend, while the dashed lines show the 95% confidence bands. The graph reveals a sharp increase in violent crime from the early 1980s to a peak in the early 1990s, followed by a significant decline through the 2000s and a leveling off in recent years. The tight confidence intervals indicate high reliability in the trend estimation, reinforcing the GAM's effectiveness in capturing historical crime dynamics.

- The Poisson GLM performs poorly, with extremely high AIC and BIC values, implying a poor fit to the data—likely because the Poisson distribution does not suit the structure of the violent crime data.

The GAM provides the best model fit and flexibility in capturing crime trends, outperforming all other models based on AIC, BIC, and deviance explained.
